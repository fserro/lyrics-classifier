```python
#this (very complicated!) loop is to add leading 0's to the filename so they keep their original order
import os
path = './songs/'
for filename in os.listdir(path):
    prefix = 'Laurie_Anderson'
    num = filename[:-4].split(sep=prefix+'_')
    num = num[1].zfill(3)
    new_filename = prefix + "_" + num + ".txt"
    src ='./songs/'+ filename
    dst ='./songs/'+ new_filename
    os.rename(src, dst)
```
